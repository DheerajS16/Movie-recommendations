```python
import numpy as np

class CollaborativeFiltering:
    def __init__(self, ratings):
        self.ratings = ratings
        self.n_users, self.n_items = ratings.shape

    def similarity(self):
        # Compute the cosine similarity between items
        item_similarity = np.zeros((self.n_items, self.n_items))
        for i in range(self.n_items):
            for j in range(self.n_items):
                if i != j:
                    # Compute cosine similarity between items i and j
                    item_similarity[i, j] = np.dot(self.ratings[:, i], self.ratings[:, j]) / (
                            np.linalg.norm(self.ratings[:, i]) * np.linalg.norm(self.ratings[:, j]))
        return item_similarity

    def recommend(self, user_id, item_similarity, n=5):
        # Get items the user has already rated
        rated_items = set(np.where(self.ratings[user_id] != 0)[0])
        
        # Compute the weighted sum of ratings for each item
        weighted_sum = np.dot(self.ratings[user_id], item_similarity)
        
        # Get the indices of items with highest weighted sum
        recommended_indices = np.argsort(weighted_sum)[::-1]
        
        # Filter out items the user has already rated
        recommended_items = [idx for idx in recommended_indices if idx not in rated_items]
        
        # Return top n recommended items
        return recommended_items[:n]

# Sample user-item ratings matrix (replace with your dataset)
ratings = np.array([
    [5, 3, 0, 1, 4],
    [4, 0, 0, 1, 4],
    [1, 1, 0, 5, 5],
    [0, 0, 5, 2, 2],
    [0, 0, 4, 4, 4],
    [0, 0, 4, 0, 0],
])

# Create collaborative filtering object
cf = CollaborativeFiltering(ratings)

# Compute item-item similarity
item_similarity = cf.similarity()

# Recommend movies for user 0
user_id = 1
recommendations = cf.recommend(user_id, item_similarity)

# Print recommended items
print("Recommended items for user", user_id, ":")
for idx in recommendations:
    print("Item", idx)
```
